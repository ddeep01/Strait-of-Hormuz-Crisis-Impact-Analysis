```python
import yfinance as yf
import pandas as pd

# Download Sensex (^BSESN)
ticker = "^BSESN"

df = yf.download(
    ticker,
    start="2025-11-01",
    end="2026-07-10",   # end date is exclusive
    interval="1d",
    auto_adjust=False
)

# Keep only required columns
df = df.reset_index()

df = df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']]

# Round prices
price_cols = ['Open', 'High', 'Low', 'Close']
df[price_cols] = df[price_cols].round(2)

# Save CSV
df.to_csv("Sensex_Daily_Mar_to_Jun_2026.csv", index=False)

print(df.head())
print(f"\nDownloaded {len(df)} rows.")
```

```python
import yfinance as yf
import pandas as pd

# Download India VIX
vix = yf.download(
    "^INDIAVIX",
    start="2025-11-01",
    end="2026-07-10",
    interval="1d",
    auto_adjust=False
)

# Reset index
vix.reset_index(inplace=True)

# Save to CSV
vix.to_csv("India_VIX_Daily.csv", index=False)

print(vix.head())
print("Saved successfully!")
```
